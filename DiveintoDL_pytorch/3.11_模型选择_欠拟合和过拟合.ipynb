这一节基本上都是一些基本概念，这里进行一些理解和总结

## 训练误差
是在训练数据集上表现出的误差吗，就是我们的项目里面运行后的train loss  
计算训练误差要用到损失函数：like：MSE，CrossEntropyLoss等
## 泛化误差
纸模型在任意一个测试数据样本上表现出的误差的期望（通过测试数据集上的误差来近似）  
同上，用到损失函数
## 对比
“训练误差”可以看作是高中平时做题（目的是为了训练）时的错误率，“泛化误差”则可以用真正参加高考时（目的是为了检测训练的程度）的错误率来近似。两个并不相同，但有时会很近似，但是并不是说训练误差很小，泛化误差就一定小，也就是（平时考试成绩很高，高考成绩也不一定高），所以“无法从训练误差估计泛化误差”  
由于我们构建模型的目的，就是为了让模型在面对遇到没有遇到过的数据时，推理能力仍然很好，所以我们更关注的时
**“泛化误差”**，但是又因为
**“无法从训练误差估计泛化误差”** ，所以并不是说我们一味全力去降低寻“训练误差”就好，我们在构建模型和训练模型时应该关注的是
**“降低泛化误差”**

## 模型选择
模型选择这一部分要先认识一些新的概念
###  验证数据集（validation dataset）
可以看到在我的projects目录下的项目中，训练模型时可以直观看到训练acc，验证acc，分别是train acc和val acc，val acc就是模型在验证数据集上的精度。  
val的目的是为了我们能够预留一小部分数据来验证我们的模型在遇到训练集之外的数据时的好坏程度，可以当作是泛化误差的前瞻（只是能反映，acc和误差当然是不一样的），一般来说得到一个数据集后，我们会先对这个数据集进行分割，分成训练集和测试集，现在引入验证集之后，就意味着仍然要分割数据集，由于数据不好获取，而且如果数据集经过不断分割，可能会倒是训练集的数据不充足，反而会舍本逐末，所以，一般情况下，我们的事件中验证数据集和测试数据集的界限比较模糊。
**实验中大多数情况下，使用的测试集在严格情况下应该是验证集。**
### k折交叉验证
上面一块提到数据集分割为训练集和验证集，如何分割？  
在训练数据不充足的情况下，把训练数据集分成k个补充和的子数据集，并作k次模型训练和验证，每一次训练，我们使用k-1个子数据集来训练模型，剩下一个子数据集来验证模型。并且在这k次训练和验证中，都是随机分割，（每次的子数据集都不相同），最后对这k次训练误差和验证误差求平均值
## 过拟合和欠拟合
这里略过，一般来说过拟合比较常见，欠拟合产生的原因大多数都是模型不够复杂，二过拟合则是模型太过复杂（参数过于多）
因为这两个问题的存在，所以我们要选择合适的模型来minimize这两个情况
